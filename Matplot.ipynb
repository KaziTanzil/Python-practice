```python
import numpy as np
X=np.arange(-10,10,0.01)
x1=np.sin(X)
y1=np.cos(X)
plt.figure(figsize=(10,1) )

plt.plot(X,x1)
plt.plot(X,y1)
plt.legend(["sin","cos"])
plt.show()
```

```python
plt.figure(figsize=(15,6))
plt.subplot(1,2,1)
plt.plot(X,x1)
plt.legend(["Sin"])
plt.subplot(1,2,2)
plt.plot(X,y1)
plt.legend(["cos"])
plt.savefig("test.pdf",transparent=True)
plt.show()

```
